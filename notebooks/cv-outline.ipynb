```python
import cv2
from matplotlib import pyplot as plt
```

## Method 1

```python
src = cv2.imread("../static/02.jpg", cv2.IMREAD_COLOR)
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
```

```python
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
binary = cv2.bitwise_not(binary)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
```

```python

for i in range(len(contours)):
    cv2.drawContours(src, [contours[i]], 0, (0, 0, 255), 2)
    cv2.putText(src, str(i), tuple(contours[i][0][0]), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 0), 1)
    print(i, hierarchy[0][i])
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```

## Method 2

```python
src = cv2.imread("../static/02.jpg", cv2.IMREAD_COLOR)
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Edge detection
edges = cv2.Canny(blurred, 50, 150)

# Finding contours
contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Optionally draw contours on the original image to visualize
cv2.drawContours(src, contours, -1, (0, 255, 0), 2)

for each in contours:
    x, y = each[0][0]
    src = cv2.putText(
        src, f'{x} {y}', (x, y), cv2.FONT_HERSHEY_SIMPLEX, .4, (0, 0, 255), 1, cv2.LINE_AA
    )

plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```
